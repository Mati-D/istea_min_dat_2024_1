```python
class Libro:
    def __init__(self, titulo, autor, genero, puntuacion):
        self.titulo = titulo
        self.autor = autor
        self.genero = genero
        self.puntuacion = puntuacion

# Crear la lista de libros
lista_libros = [
    Libro("Cien años de Soledad", "Gabriel García Márquez", "Ficcion", 4.5),
    Libro("1984", "George Orwell", "Ficcion", 4.3),
    Libro("El Hobbit", "J.R.R. Tolkien", "Fantasia", 4.7),
    Libro("Orgullo y Prejuicio", "Jane Austen", "Romance", 4.2),
    Libro("Crimen y Castigo", "Fiodor Dostoyevsky", "Clasico", 4.4),
    Libro("Los Juegos del Hambre", "Suzanne Collins", "Juvenil", 4.1),
    Libro("Don Quijote de la Mancha", "Miguel de Cervantes", "Clasico", 4.6),
    Libro("Harry Potter y la Piedra Filosofal", "J.K. Rowling", "Fantasia", 4.8),
    Libro("Los pilares de la tierra", "Ken Follet", "Historica", 4.4),
    Libro("Cazadores de sombras: ciudad de hueso", "Cassandra Clare", "Fantasia", 4.4)
]

```

```python
# Función para agregar un nuevo libro a la lista
def agregar_libro():
    titulo = input("Ingrese el título del libro: ")
    autor = input("Ingrese el nombre del autor: ")
    genero = input("Ingrese el género del libro: ")
    puntuacion = float(input("Ingrese la puntuación del libro: "))
    libro_nuevo = Libro(titulo, autor, genero, puntuacion)
    lista_libros.append(libro_nuevo)
    print("Libro agregado correctamente.")

```

```python
# Función para buscar libros por género
def buscar_por_genero():
    genero_buscar = input("Ingrese el género que desea buscar: ")
    print("Libros encontrados del género", genero_buscar, ":")
    for libro in lista_libros:
        if libro.genero.lower() == genero_buscar.lower():
            print(libro.titulo)

```

```python
# Función para recomendar un libro del género con la puntuación más alta
def recomendar_libro():
    genero_interes = input("Ingrese su género de interés: ")
    max_puntuacion = 0
    libro_recomendado = None
    for libro in lista_libros:
        if libro.genero.lower() == genero_interes.lower() and libro.puntuacion > max_puntuacion:
            max_puntuacion = libro.puntuacion
            libro_recomendado = libro
    if libro_recomendado:
        print("¡Recomendación!")
        print("Título:", libro_recomendado.titulo)
        print("Puntuación:", libro_recomendado.puntuacion)
    else:
        print("Lo siento, no se encontraron libros en ese género.")

```

```python
# Bucle principal
while True:
    print("\n--- Menú ---")
    print("1. Agregar libro")
    print("2. Buscar libros por género")
    print("3. Recomendar libro")
    print("4. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        agregar_libro()
    elif opcion == "2":
        buscar_por_genero()
    elif opcion == "3":
        recomendar_libro()
    elif opcion == "4":
        print("¡Hasta luego!")
        break
    else:
        print("Opción no válida. Por favor, seleccione una opción válida.")
```

```python
# Mostrar información de los libros en la lista
for libro in lista_libros:
    print(f"Título: {libro.titulo}")
    print(f"Autor: {libro.autor}")
    print(f"Género: {libro.genero}")
    print(f"Puntuación: {libro.puntuacion}")
    print()
```
